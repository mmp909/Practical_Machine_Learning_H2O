Merge: use h2o.merge() to join tables together when they have 1+ columns in common.

```python
A = h2o.H2OFrame(
    pd.DataFrame({
        'a':range(1,11),
        'b':range(11,21),
        'c':range(21,31)
    }),
destination_frame = "A"
)
```

```python
A.as_data_frame()
```

```python
np.random.seed(123)
B = h2o.H2OFrame(
    pd.DataFrame({
        'a':np.random.randint(6,13,20),
        'b':np.random.randint(6,13,20)
    }),
destination_frame = "B"
)
```

```python
B.as_data_frame()
```

```python
M = A.merge(B)
```

```python
M.as_data_frame()
```

```python
M = B.merge(A)
```

```python
M.as_data_frame()
```

```python
M = B.merge(A, by_x="b", by_y=["a"])
```

```python
M.as_data_frame()
```

```python
M = B.merge(A, by_y="b", by_x=["a"])
```

```python
M.as_data_frame()
```

```python
M = B.merge(A, by_y="b", by_x=["a"])
```

```python
M.as_data_frame()
```

```python
M = A.merge(B, by_y="b", by_x=["a"])
```

```python
M.as_data_frame()
```

```python
M = A.merge(B, by_y="a", by_x=["a"], all_x=True)
```

```python
M.as_data_frame()
```
